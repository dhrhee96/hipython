```python
#재고관리
inventory= {'노트북':0,
            '키보드':0,
            '마우스':120,
            '모니터':15}
#값=input('넣으세요')
#amt=inventory[값]
for name,amt in inventory.items():
    if amt >50 :
        status= ("충분")
    elif amt>10 :
        status= ("적정")
    elif amt >0 :
        status=("부족")
    else:
        status="품절"
    print(f'{name} 는 {amt}로 {status} 하다')
    
```

```python
sentence = input("문장을 입력하세요: ")
words = ["파이썬", "코딩", "데이터", "AI"]

found_word = None

for word in words:
    if word in sentence:
        found_word = word
        break

if found_word:
    print(f"입력한 문장에서 '{found_word}' 단어를 찾았습니다!")
else:
    print("입력한 문장에 해당 단어가 없습니다.")
```

```python
numbers = [1, 3, 5, 7, 9, 11, 13, 15]
search_number = 9
idx= 1

for num in numbers:
    if num == search_number:
        print(f"{idx} 번째서 찾았습니다!")
        break
    print(f"현재 검사 중인 숫자: {num}")
    idx= idx+1
```
