```python
person = {"이름": "Alice", "나이": 25}
person.update({"나이": 26, "직업": "개발자"})
print(person)  # {'이름': 'Alice', '나이': 26, '직업': '개발자'}
```

```python
print(person.get("직업", "정보 없음"))  # "정보 없음"

d = {"a":1, "b":2}
print( list(d.keys()) )
print( list(d.values()) )
for k,v in d.items() :
	print(k,v)

print(d.get("x",0) )
```

```python
person = {"이름": "Alice", "나이": 25, "도시": "서울"}

# 값 추가
person["직업"] = "개발자"

# 값 수정
person["나이"] = 26

# 키 삭제
del person["도시"]
```

```python
print(person.keys())     # dict_keys(['이름', '나이', '직업'])
print(person.values())   # dict_values(['Alice', 26, '개발자'])
print(person.items())    # dict_items([('이름', 'Alice'), ('나이', 26), ('직업', '개발자')])
print(person.pop("나이"))# 26
```

```python
A = {1, 2, 3, 4}
B = {3, 4, 5, 6}

print(A | B)  # {1, 2, 3, 4, 5, 6} (합집합)
print(A & B)  # {3, 4} (교집합)
print(A - B)  # {1, 2} (차집합)
```
