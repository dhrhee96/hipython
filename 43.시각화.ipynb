# 시각화 라이브러리 설치

```python
import seaborn as sns
anscombe = sns.load_dataset('anscombe')
anscombe
```

```python
import warnings
warnings.filterwarnings('ignore')
#폰트 지정
from matplotlib import rcParams
rcParams['font.family'] = 'Gulim'
rcParams['axes.unicode_minus'] = False
import matplotlib.pyplot as plt
plt.title('클래스와 오브젝트')
plt.xlabel('클래스')
plt.ylabel('오브젝트')
```

```python
sns.get_dataset_names()
```

```python
d1 = anscombe[anscombe['dataset']=='I']
d2 = anscombe[anscombe['dataset']=='II']
d3 = anscombe[anscombe['dataset']=='III']
d4 = anscombe[anscombe['dataset']=='IV']
d1.describe()
d2.describe()
d3.describe()
d4.describe()
```

```python
fig = plt.figure()
axes1 = fig.add_subplot(2,2,1) #2row, 2col,position
axes2 = fig.add_subplot(2,2,2)
axes3 = fig.add_subplot(2,2,3)
axes4 = fig.add_subplot(2,2,4)

axes1.plot( d1['x'], d1['y'] , 'o')
axes2.plot( d2['x'], d2['y'] , 'o')
axes3.plot( d3['x'], d3['y'] , 'o')
axes4.plot( d4['x'], d4['y'] , 'o')

axes1.set_title('dataset I')
axes2.set_title('dataset II')
axes3.set_title('dataset III')
axes4.set_title('dataset IV')

fig.suptitle('Anscombe Dataset 으로 보는 데이터 시각화의 의미')
fig.set_tight_layout(True)
plt.show()
```

```python
dates = ['2024-10-01', '2024-10-02', '2024-10-03', '2024-10-04', '2024-10-05']
prices = [100, 102, 101, 105, 103]

# 단일 그래프 설정 (빨간색 테두리 영역)
plt.figure(figsize=(10, 6))
plt.title("Stock Prices Over 5 Days")
plt.xlabel("Date")
plt.ylabel("Price")
plt.plot(dates, prices, marker='o', label='Stock Price', color='blue')
plt.legend()
plt.grid()
plt.ylim(95, 110)
plt.xlim(0, 4)
```

```python
plt.subplot(1, 2, 1)
plt.plot(dates, prices, marker='o', color='green')
plt.title("Stock Prices - Subplot 1")
plt.xticks(rotation=45)
```

```python
plt.subplot(1, 2, 2)
plt.bar(dates, prices, color='orange')
plt.title("Stock Prices - Subplot 2")
plt.xticks(rotation=45)

# 결과 출력 및 저장
plt.show()
```

# 용도에 맞는 데이터 그리기 

```python
tips_df= sns.load_dataset('tips')
tips_df
```

# 1변량 그래프: 히스토그램 - 구간, 빈도수

```python
plt.hist(data=tips_df,x='total_bill',bins=10,rwidth=0.8)
```

# total bill vs tip

```python
plt.subplot(1,2,1)
plt.hist(data=tips_df,x='total_bill',bins=10,rwidth=0.8)
plt.subplot(1,2,2)
plt.hist(data=tips_df,x='tip',bins=10,rwidth=0.8)
```

# 1변량 그래프 : 박스플롯

```python
plt.boxplot(
    [tips_df['total_bill'], tips_df['tip']],
    labels=['total_bill', 'tip']
)
plt.grid()
plt.show()
```

```python
# 남성 vs 여성
plt.boxplot(
    [tips_df[tips_df['sex']=='Male']['total_bill'], tips_df[tips_df['sex']=='Female']['total_bill']],
    labels=['Male', 'Female']
)
plt.grid()
plt.show()
```

```python
tips_df.groupby(['day','time']).size()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=tips_df, x='day', y='total_bill', hue='time')

# 그래프 제목 및 라벨 설정
plt.title('Total Bill Distribution by Day and Time')
plt.xlabel('Day of the Week')
plt.ylabel('Total Bill ($)')
plt.legend(title='Time')

# 그래프 표시
plt.show()
```
