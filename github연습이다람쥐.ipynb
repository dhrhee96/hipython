```python
# Í≥ÑÏÇ∞Îêú Í∞íÏùÑ Ï†ÄÏû•Ìï† Ï£ºÎ®∏Îãà(ÎîïÏÖîÎÑàÎ¶¨)
memo = {0: 0, 1: 1}

def fibo_memo(n):
    # Ïù¥ÎØ∏ Í≥ÑÏÇ∞Ìïú Ï†ÅÏù¥ ÏûàÎã§Î©¥ Ï£ºÎ®∏ÎãàÏóêÏÑú Í∫ºÎÇ¥Ï§å
    if n in memo:
        return memo[n]
    
    # Ï≤òÏùå Í≥ÑÏÇ∞ÌïòÎäî Í±∞ÎùºÎ©¥ Í≥ÑÏÇ∞ ÌõÑ Ï£ºÎ®∏ÎãàÏóê Ï†ÄÏû•
    memo[n] = fibo_memo(n - 1) + fibo_memo(n - 2)
    return memo[n]

print([fibo_memo(i) for i in range(10)])
```

```python
from itertools import groupby

def compress_with_groupby(s):
    if not s:
        return ""
    
    result = ""
    # groupbyÎäî Ïó∞ÏÜçÎêú Î¨∏ÏûêÎ•º (Î¨∏Ïûê, Í∑∏Î£πÎ∞òÎ≥µÏûê) ÌòïÌÉúÎ°ú Î¨∂Ïñ¥Ï§çÎãàÎã§.
    for char, group in groupby(s):
        count = len(list(group))  # Í∑∏Î£π ÎÇ¥Ïùò Í∞úÏàòÎ•º ÏÖâÎãàÎã§.
        result += f"{char}{count}"
        
    return result

# Ïã§Ìñâ ÏòàÏãú
input_str = "aaabbcccccca"
print(f"Ï∂úÎ†•: {compress_with_groupby(input_str)}") # a3b2c6a1
```

```python
def check_all_digits(s):
    # 1. Í∏∏Ïù¥Í∞Ä Ï†ïÌôïÌûà 10Ïù¥Ïñ¥Ïïº Ìï® (0~9ÍπåÏßÄ Ï¥ù 10Í∞ú)
    if len(s) != 10:
        return False
    
    # 2. setÏúºÎ°ú ÎßåÎì§ÏóàÏùÑ ÎïåÎèÑ Í∏∏Ïù¥Í∞Ä 10Ïù¥Î©¥ Ï§ëÎ≥µÏù¥ ÏóÜÎã§Îäî Îúª
    return len(set(s)) == 10

# ÌÖåÏä§Ìä∏
print(check_all_digits("0123456789")) # True
print(check_all_digits("0123445678")) # False (Ï§ëÎ≥µ)
print(check_all_digits("12345"))      # False (Í∏∏Ïù¥ ÎØ∏Îã¨)
print(check_all_digits("6789012345")) 
print(check_all_digits("012322456789")) 
```
