# 자동차 클래스 정의

-self.속성명:인스턴스 속성
-속성명:클래스 속성

- __init__(self) 생성자에서 => 인스턴스 속성 설정
- def 메소드(self): 메소드 정의
-정적 매소드 : 유틸리티
-클래스 메소드(cls) : 클래스 속성 강의,@classmethod => "decorator"

```python
class Car: 
  #1- 클래스 속성
  instance_cnt=0

  #def __init__(self):
   #  print('자동차 객체가 생성되었다이')
  def __init__(self,color=''):
    # 객체의 속성
     self.color=color
     
     Car.instance_cnt +=1

     print(f'{self.color} 자동차가 움직입니다.{Car.instance_cnt} 대')
  def auto_pilot(self):
    print(f'{self.color} 자동차가 자율주행중이므니다')
    print(f' {Car.instance_cnt} 대 자동차가 움직입니다.')
  def count_check(cls):
    print(f'자동차 대수:{cls.instance_cnt}')
  @staticmethod 
  def type_check(type_code): # 유틸리티
    if type_code > 20:
       print('전기차')
    elif type_code < 20:
      print('내연자동차')
    else:
      print('하이브리드')    
```

```python
class Calculator:
    def __init__(self, numbers):
        # 숫자를 리스트 형태로 저장합니다.
        self.numbers = numbers

    def sum(self):
        total = 0
        for num in self.numbers:
            total += num
        return total

    def avg(self):
        # 리스트가 비어있을 경우 0으로 나누는 에러를 방지합니다.
        if not self.numbers:
            return 0
        return self.sum() / len(self.numbers)

```

```python
cal1=Calculator([1,2,3,4,5])
cal2=Calculator([6,7,8,9,10])
```

```python
cal1.sum(),cal1.avg(),cal2.sum(),cal2.avg()
```

```python
# 정적방법이다람쥐
class Calculator:
    @staticmethod
    def sum(numbers):
        """숫자 리스트의 합계를 반환합니다."""
        return sum(numbers)

    @staticmethod
    def avg(numbers):
        """숫자 리스트의 평균을 반환합니다. 비어있으면 0을 반환합니다."""
        if not numbers:
            return 0
        return sum(numbers) / len(numbers)
nums=[6,7,8,9,10]
# 사용 예시 (객체 생성 없이 바로 사용 가능)
print(f"합계: {Calculator.sum(nums)}")
print(f"평균: {Calculator.avg(nums)}")
```

# Q2. Account 클래스 은행에 가서 계좌를 개설하면 은행이름, 예금주, 계좌번호, 잔액이 설정됩니다. Account 클래스를 생성한 후 생성자를 구현해보세요. 생성자에서는 예금주와 초기 잔액만 입력 받습니다. 은행이름은 SC은행으로 계좌번호는 3자리-2자리-6자리 형태로 랜덤하게 생성됩니다.

은행이름: SC은행 계좌번호: 111-11-111111

```python
import random

class Account:
    def __init__(self, owner, balance):
        self.bank = "SC은행"
        self.owner = owner
        self.balance = balance
        
        # 3자리-2자리-6자리 랜덤 계좌번호 생성
        n1 = random.randint(0, 999)
        n2 = random.randint(0, 99)
        n3 = random.randint(0, 999999)
        self.account_number = f"{n1:03d}-{n2:02d}-{n3:06d}"

# 실행 예시
acc = Account("김파이", 10000)
print(f"은행: {acc.bank} / 예금주: {acc.owner} / 계좌번호: {acc.account_number} / 잔액: {acc.balance}원")
```

```python
class Animal:
    def make_sound(self):
        pass

class Dog(Animal):
    def make_sound(self):
        return "멍멍!"

class Cat(Animal):
    def make_sound(self):
        return "야옹!"
```

```python
animals = [Dog(), Cat()]
for animal in animals:
    print(animal.make_sound())  # "멍멍!", "야옹!" 출력
```
