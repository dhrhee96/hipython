```python
students = [{'num':'1','name':'kim', 'kor':80,'math':90,'eng':100,},
            {'num':'2','name':'park', 'kor':90,'math':60,'eng':80,},
            {'num':'3','name':'hong', 'kor':70,'math':70,'eng':80,},
            {'num':'4','name':'hong1', 'kor':70,'math':70,'eng':80,},
            {'num':'5','name':'hong2', 'kor':70,'math':70,'eng':80,},
           ]
```

```python
# 총점,평균,순위-> 속성추가 -> 출력
# 전체 총점 평균 출력
# 과목별 평균 출력
students = [
    {'num':'1','name':'kim', 'kor':80,'math':90,'eng':100},
    {'num':'2','name':'park', 'kor':90,'math':60,'eng':80},
    {'num':'3','name':'hong', 'kor':70,'math':70,'eng':80},
    {'num':'4','name':'hong1', 'kor':70,'math':70,'eng':80},
    {'num':'5','name':'hong2', 'kor':70,'math':70,'eng':80},
]

# 1. 개인별 총점 및 평균 속성 추가
for s in students:
    s['total'] = s['kor'] + s['math'] + s['eng']
    s['avg'] = s['total'] / 3

# 2. 전체 총점 및 전체 평균 계산
total_sum = 0
for s in students:
    total_sum += s['total']

total_avg = total_sum / (len(students) * 3)

print("-" * 55)
print(f"전체 학생 총점 합계: {total_sum}점")
print(f"전체 학생 평균 점수: {total_avg:.2f}점")

# 3. 과목별 평균 계산
kor_total = 0
math_total = 0
eng_total = 0

for s in students:
    kor_total += s['kor']
    math_total += s['math']
    eng_total += s['eng']

num_of_students = len(students)
print("\n--- [과목별 평균] ---")
print(f"국어 평균: {kor_total / num_of_students:.1f}점")
print(f"수학 평균: {math_total / num_of_students:.1f}점")
print(f"영어 평균: {eng_total / num_of_students:.1f}점")
```

# 업앤 다운 게임

```python
import random
number = random.randint(1,10)
number
```

```python
# 정답
cnt=0
while True: 
    # 입력
    answer=int(input("1-10 사이까지 넣어라"))
    cnt +=1
    if answer == number:
       print(f"앙 축하띠,도전횟수{cnt}")
       break 
    elif answer > number:
       print('down')
    else:
       print('up')
```
