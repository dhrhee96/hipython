# 변수의 영역 namespace

```python
def local_exam(): # 사용자 정의 함수
    local_var=100
    print(f"함수내부:{local_var}")

local_exam()
local_var=200
local_exam()
print(f"함수외부:{local_var}")
```

```python
counter=0
def inc_counter():
     global counter
     counter +=1
     print(counter)
inc_counter()
inc_counter()
inc_counter()
```
