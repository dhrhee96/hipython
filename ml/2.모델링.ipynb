# k-fold cross validation 교차검증

```python
import pandas as pd
iris_data_df = pd.DataFrame(data=iris_data, columns=iris.feature_names)
iris_data_df
```

```python
iris_data_df['label']=iris.target
iris_data_df['label'].value_counts()
```

```python
from sklearn.model_selection import KFold
kfold_5=KFold(n_splits=5)
kfold_5
```

```python
iris_cv_clf = DecisionTreeClassifier(random_state=42)
```

```python
for train_idx, test_idx in kfold_5.split(iris_data_df):
    X_train, X_test = iris_data_df.iloc[train_idx], iris_data_df.iloc[test_idx]
    y_train, y_test = iris_data_df['label'].iloc[train_idx], iris_data_df['label'].iloc[test_idx]
    
    iris_cv_clf.fit(X_train, y_train)
    iris_cv_pred = iris_cv_clf.predict(X_test)
    print(accuracy_score(y_test, iris_cv_pred))
```

# 프로세스 연습

```python
from sklearn.datasets import load_wine

# ==========================================
# 1. 데이터 준비
# ==========================================
wine = load_wine()

# X: 문제 (알코올 농도, 색상 등 13가지 화학적 특성)
# y: 정답 (0, 1, 2로 나누어진 3가지 와인 품종)
X = wine.data
y = wine.target

# 전체 데이터를 학습용(80%)과 테스트용(20%)으로 분리
# random_state를 지정하면 실행할 때마다 똑같이 섞이게 됩니다.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=38)

# ==========================================
# 2. 모델 생성 및 학습
# ==========================================
# 의사결정나무 알고리즘 구현 클래스 생성 (깡통)
model = DecisionTreeClassifier(random_state=38)

# 학습용 문제와 정답을 주고 모델 학습시키기
model.fit(X_train, y_train)

# ==========================================
# 3. 예측 및 평가
# ==========================================
# 학습에 사용하지 않은 테스트용 문제(X_test)를 주고 예측해보기
y_pred = model.predict(X_test)

# 실제 정답(y_test)과 모델이 예측한 값(y_pred)을 비교하여 정확도 계산
accuracy = accuracy_score(y_test, y_pred)

print("=== 와인 품종 분류 결과 ===")
print(f"모델 정확도 : {accuracy * 100:.2f}%")
```

```python
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier
import pandas as pd

# 1. 데이터 로드
wine = load_wine()
X_train, X_test, y_train, y_test = train_test_split(
    wine.data, wine.target, test_size=0.2, random_state=156
)

# 2. 데이터 스케일링 (StandardScaler 적용)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test) # 테스트 데이터는 fit 없이 transform만!

# 3. GridSearchCV 설정
wine_dt_gcv = DecisionTreeClassifier(random_state=156)
params = {
    'max_depth': [1, 2, 3],
    'min_samples_split': [2, 3, 4]
}

# cv_results_를 보기 쉽게 출력하기 위해 refit=True 설정 (기본값)
grid_cv = GridSearchCV(wine_dt_gcv, param_grid=params, cv=5, scoring='accuracy')

# 4. 학습
grid_cv.fit(X_train_scaled, y_train)

# 5. 결과 확인 (results_ 가 아니라 cv_results_ 입니다)
# 데이터를 표 형태로 보기 위해 DataFrame으로 변환합니다.
results_df = pd.DataFrame(grid_cv.cv_results_)
print(results_df[['params', 'mean_test_score', 'rank_test_score']])

print(f"\n최적 하이퍼 파라미터: {grid_cv.best_params_}")
print(f"최고 정확도: {grid_cv.best_score_:.4f}")
```

모델링 프로세스는 동일하다. 
- 각 모델의 하이퍼 파라미터가 다르다.

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score


models = {
    'KNN': KNeighborsClassifier(),
    'RF': RandomForestClassifier(),
    'LR': LogisticRegression()
}
for name, model in models.items():
    model.fit(X_train_scaled, y_train)
    pred = model.predict(X_test_scaled)
    acc = accuracy_score(y_test, pred)
    print(f"{name} 정확도: {acc:.4f}")
```
