```python
# * 가변인자
def 계산기(*a):
    print(a)
    return sum(a)
```

```python
def calc(a,b,*c):
    sum_all=a+b+sum(c)
    return sum_all
```

```python
calc(2,3,4,5,6)
```

## 간식박스

```python
함수도출 : CRUD
함수구현
문제점 도출 
전체흐름 구현 함수
```

```python
# 간식박스 생성 함수
def make_snackbox():
    snackbox = ['두쫀쿠']
    return snackbox
```

```python
# 간식박스 추가함수
def add_snackbox(sbox): #간식리스트
    make_msg = input('먹고 싶은 간식을 추가하세요. 쉼표(,)로 연결하세요').split(',')
    sbox.extend(make_msg)
    #print(sbox)
    return sbox
```

```python
def display_snackbox(sbox):
    print(sbox)
```

```python
# 메뉴출력
def display_menu():
    print("1. 간식 추가")
    print("2. 간식박스 확인")
    print("3. 간식 제거")
    print("4. 주문 획정")
```

```python
def order_confirm():
    make_msg=input('주문을 확정하겠습니다?(yes/no)')
    if make_msg== 'yes'
       return True
    else : 
      
    pass
```

```python
#서비스의 흐름 로직
snack_box = make_snackbox()
print('간식박스 생성완료', snack_box)

while True:
    display_menu() #
    menu_no = input('메뉴를 선택하세요')
    if menu_no == '1':
        add_snackbox(snack_box) #간식리스트
    elif menu_no == '2':
        display_snackbox(snack_box)
    elif menu_no == '4':
        display_snackbox(snack_box)
        result = order_confirm() #매개변수 통일
        if result :
            break
print('고객님의 주문을 감사드립니다')
```
