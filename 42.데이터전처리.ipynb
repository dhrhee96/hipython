# merge()
- 키를 기준으로 합친다
- sql 에서 join

```python
import pandas as pd
```

```python
# 매출데이터
sales_data = pd.DataFrame({ 'CustomerID': [101, 102, 103, 104], 'Date': ['2023-01-01', '2023-01-03', '2023-01-05', '2023-01-07'], 'Amount': [250, 300, 150, 400] })
# 고객데이터
customer_data = pd.DataFrame({ 'CustomerID': [101, 102, 103, 105], 'Name': ['Alice', 'Bob', 'Charlie', 'Eva'], 'Region': ['North', 'South', 'East', 'West'] })
```

```python
sales_data.info()
customer_data.info()
```

```python
pd.merge(customer_data,sales_data,on='CustomerID')
```

```python
pd.merge(customer_data,sales_data,on='CustomerID',how='right')
```

how- 병합 기준 지정
- left
- right 
- inner(default)
- outer 
- on(병합키 지정)

# 결측치처리 

```python
import numpy as np
pd.NA,np.nan
```

```python
pd.isnull(np.nan),pd.isnull(pd.NA)
```

```python
data = {
    'name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'age': [25, np.nan, 30, 22, np.nan],
    'score': [90, 85, np.nan, 88, 95],
    'city': ['Seoul', 'Busan', np.nan, 'Incheon', 'Seoul']
}
data
```

```python
df = pd.DataFrame(data)

# 결측치 채우기 및 포맷 정리
df['age'] = df['age'].fillna(df['age'].mean()).round(1)
df['score'] = df['score'].fillna(df['score'].mean()).round(1)
df['city'] = df['city'].fillna('Unknown')
```
