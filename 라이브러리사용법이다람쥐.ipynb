```python
import numpy as np

# 초기 매출
initial = 100

# 시나리오 적용
step1 = initial * 2.0      # +100%
step2 = step1 * 0.5        # -50%
final = step2

print(f"초기: {initial}만원")
print(f"1차 후: {step1}만원")
print(f"최종: {final}만원")
print(f"\n실제 변화: {((final/initial - 1) * 100):.1f}%")

# 산술평균
arithmetic = (100 + (-50)) / 2
print(f"산술평균: {arithmetic:.1f}%")

# 기하평균
geometric = (np.sqrt(2.0 * 0.5) - 1) * 100
print(f"기하평균: {geometric:.1f}%")
```

```python
import numpy as np
from scipy import stats

# 1. 데이터: 갈 때 100km/h, 올 때 50km/h
speeds = [100, 50]

# 2. 평균 계산 비교
# A. 산술평균 (오답)
arithmetic = np.mean(speeds)

# B. 조화평균 (정답)
harmonic = stats.hmean(speeds)

# 3. 결과 리포트
print(f"데이터: {speeds}")
print("-" * 35)
print(f"1. [❌ 오류] 산술평균: {arithmetic:.2f} km/h")
print("-" * 35)
print(f"2. [✅ 정답] 조화평균: {harmonic:.2f} km/h")
print(f"   >> 해석: '느린 구간(50) 때문에 전체 효율이 급감했습니다.'")
print(f"   >> 실제 물리적 평균 속도는 66.67km/h가 맞습니다.")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import kurtosis, norm, laplace, uniform

# --- 1. Data Generation ---
# Let's create a few different datasets to demonstrate kurtosis
np.random.seed(42) # for reproducibility

# Normal distribution (Mesokurtic - Kurtosis close to 0 for Fisher's, 3 for Pearson's)
data_normal = np.random.normal(loc=0, scale=1, size=1000)

# Laplace distribution (Leptokurtic - Heavier tails than normal)
data_laplace = laplace.rvs(loc=0, scale=1, size=1000)

# Uniform distribution (Platykurtic - Lighter tails than normal, "flat")
data_uniform = uniform.rvs(loc=-2, scale=4, size=1000) # From -2 to 2

# Custom data with outliers to show high kurtosis
data_outliers = np.concatenate([np.random.normal(0, 1, 950),
                                np.random.normal(0, 10, 50)])

# --- 2. Kurtosis Calculation ---
# SciPy's kurtosis function calculates Fisher's kurtosis (excess kurtosis),
# where a normal distribution has a kurtosis of 0.

print("--- Kurtosis Values (Fisher's Excess Kurtosis) ---")
print(f"Normal Distribution Kurtosis: {kurtosis(data_normal):.4f}")
print(f"Laplace Distribution Kurtosis: {kurtosis(data_laplace):.4f}")
print(f"Uniform Distribution Kurtosis: {kurtosis(data_uniform):.4f}")
print(f"Outlier Data Kurtosis: {kurtosis(data_outliers):.4f}")


# --- 3. Visualizations ---

# Histogram and KDE to compare distributions
plt.figure(figsize=(15, 10))

# Normal Distribution
plt.subplot(2, 2, 1)
plt.hist(data_normal, bins=30, density=True, alpha=0.6, color='skyblue', label='Normal Data')
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, np.mean(data_normal), np.std(data_normal))
plt.plot(x, p, 'k', linewidth=2, label='Normal PDF')
plt.title(f'Normal Distribution (Kurtosis: {kurtosis(data_normal):.2f})')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)

# Laplace Distribution (Leptokurtic)
plt.subplot(2, 2, 2)
plt.hist(data_laplace, bins=30, density=True, alpha=0.6, color='lightcoral', label='Laplace Data')
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = laplace.pdf(x, np.mean(data_laplace), np.std(data_laplace, ddof=1)) # ddof=1 for sample std dev
plt.plot(x, p, 'k', linewidth=2, label='Laplace PDF')
plt.title(f'Laplace Distribution (Kurtosis: {kurtosis(data_laplace):.2f})')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)

# Uniform Distribution (Platykurtic)
plt.subplot(2, 2, 3)
plt.hist(data_uniform, bins=30, density=True, alpha=0.6, color='lightgreen', label='Uniform Data')
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = uniform.pdf(x, loc=np.min(data_uniform), scale=(np.max(data_uniform) - np.min(data_uniform)))
plt.plot(x, p, 'k', linewidth=2, label='Uniform PDF')
plt.title(f'Uniform Distribution (Kurtosis: {kurtosis(data_uniform):.2f})')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)

# Data with Outliers (High Kurtosis)
plt.subplot(2, 2, 4)
plt.hist(data_outliers, bins=30, density=True, alpha=0.6, color='plum', label='Outlier Data')
plt.title(f'Data with Outliers (Kurtosis: {kurtosis(data_outliers):.2f})')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()

# Box Plot to visualize outliers and spread, indirectly related to kurtosis
plt.figure(figsize=(10, 6))
data_to_plot = [data_normal, data_laplace, data_uniform, data_outliers]
labels = ['Normal', 'Laplace', 'Uniform', 'Outliers']
plt.boxplot(data_to_plot, labels=labels, vert=False, patch_artist=True)
plt.title('Box Plots of Different Distributions')
plt.xlabel('Value')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()

# Q-Q plot: A powerful visual tool to assess if a distribution is normal (or other specified distribution)
# Deviations from the straight line indicate non-normality, which often relates to skewness and kurtosis.
from statsmodels.graphics.gofplots import qqplot

fig, axes = plt.subplots(2, 2, figsize=(12, 10))
qqplot(data_normal, line='s', ax=axes[0, 0])
axes[0, 0].set_title(f'Normal Q-Q Plot (Kurtosis: {kurtosis(data_normal):.2f})')

qqplot(data_laplace, line='s', ax=axes[0, 1])
axes[0, 1].set_title(f'Laplace Q-Q Plot (Kurtosis: {kurtosis(data_laplace):.2f})')

qqplot(data_uniform, line='s', ax=axes[1, 0])
axes[1, 0].set_title(f'Uniform Q-Q Plot (Kurtosis: {kurtosis(data_uniform):.2f})')

qqplot(data_outliers, line='s', ax=axes[1, 1])
axes[1, 1].set_title(f'Outlier Data Q-Q Plot (Kurtosis: {kurtosis(data_outliers):.2f})')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# 1. 데이터 로드
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/housing/housing.data"
column_names = [
    'CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 
    'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV'
]
# 실제 환경에서 실행 시 아래 줄의 주석을 해제하세요
df = pd.read_csv(url, sep='\s+', names=column_names)

# 2. 다항 회귀(Polyfit) 계산
x = df['RM']
y = df['MEDV']

# 1차항(직선) 추세선 계산
z1 = np.polyfit(x, y, 1)
p1 = np.poly1d(z1)

# 2차항(곡선) 추세선 계산
z2 = np.polyfit(x, y, 2)
p2 = np.poly1d(z2)

# 추세선을 그리기 위한 X축 범위 생성
x_range = np.linspace(x.min(), x.max(), 100)

# 3. 시각화
plt.figure(figsize=(10, 6))

# 산점도
plt.scatter(x, y, alpha=0.5, color='royalblue', label='Data Points')

# 추세선 추가
plt.plot(x_range, p1(x_range), "r--", lw=2, label=f"Linear Fit (deg=1)")
plt.plot(x_range, p2(x_range), "g-", lw=2, label="Quadratic Fit (deg=2)")

# 그래프 꾸미기
plt.title('RM vs MEDV with Polynomial Regression Lines', fontsize=14)
plt.xlabel('Average Number of Rooms (RM)', fontsize=12)
plt.ylabel('Median Value of Homes (MEDV)', fontsize=12)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)

plt.show()
```

```python
import numpy as np
from scipy import stats

# ==========================================
# 1단계: 데이터 수집 (가상 데이터 생성)
# ==========================================
np.random.seed(42)  # 결과 재현을 위한 시드 설정

# 상황: 통증 수치(0~100)가 얼마나 줄어들었는가? (수치가 높을수록 효과 좋음)
# 기존 진통제(A그룹): 평균 50만큼 감소, 표준편차 10, 환자 30명
group_old = np.random.normal(loc=50, scale=10, size=30)

# 신규 진통제(B그룹): 평균 55만큼 감소, 표준편차 10, 환자 30명
# (우리는 신규 약이 더 좋다고 가정하고 데이터를 만들었습니다)
group_new = np.random.normal(loc=55, scale=10, size=30)

print(f"[데이터 확인]")
print(f"기존 진통제 효과 평균: {group_old.mean():.2f}")
print(f"신규 진통제 효과 평균: {group_new.mean():.2f}")
print("-" * 30)


# ==========================================
# 2단계: 가설 수립 및 검정 통계량 산출
# ==========================================
# H0: New <= Old (효과가 없거나 같다)
# H1: New > Old (신규 약이 더 우수하다) -> 'greater' 옵션 사용 (단측 검정)

# 독립 표본 T-검정 수행
t_stat, p_value = stats.ttest_ind(
    group_new,       # 신규 약 (대립가설의 주인공)
    group_old,       # 기존 약
    equal_var=False, # 등분산 가정 안 함 (Welch's t-test, 실무적으로 더 안전함)
    alternative='greater' # 중요: "더 크다(우수하다)"를 검증하므로 우측 검정
)

print(f"[검정 결과]")
print(f"T-statistic (차이의 크기): {t_stat:.4f}")
print(f"P-value (우연일 확률): {p_value:.4f}")
print("-" * 30)


# ==========================================
# 3단계: 최종 결정 (비즈니스 의사결정)
# ==========================================
alpha = 0.05  # 유의수준 (리스크 허용 범위 5%)

print(f"[최종 판단]")
if p_value < alpha:
    print(f"결과: 귀무가설 기각 (P-value {p_value:.4f} < {alpha})")
    print("해석: 신규 진통제의 효능이 통계적으로 유의미하게 우수합니다.")
    print("👉 Action: 신규 약 출시 승인 / 마케팅 포인트로 활용")
else:
    print(f"결과: 귀무가설 채택 (P-value {p_value:.4f} >= {alpha})")
    print("해석: 평균은 조금 높아 보일지 몰라도, 통계적으로 우수하다고 볼 수 없습니다 (우연일 가능성 높음).")
    print("👉 Action: 출시 보류 / 추가 임상 실험 필요")
```

```python
import numpy as np
import scipy.stats as stats

np.random.seed(1) # 시드 고정

# [상황 설정]
# 신(God): 실제 평균은 70kg이 아니라 '70.5kg'이다. (차이가 존재함!)
# 하지만 표준편차(산포)가 커서 데이터가 들쭉날쭉함
real_weight = np.random.normal(loc=70.5, scale=10, size=100)

# [가설 검정]
# 귀무가설: 평균은 70kg이다.
t_stat, p_value = stats.ttest_1samp(real_weight, 70)

print(f"표본 평균: {np.mean(real_weight):.2f}")
print(f"P-value: {p_value:.4f}")

if p_value >= 0.05:
    print("결과: 귀무가설 채택 (차이 없다)")
    print("👉 오류 발생!: 실제로는 70.5kg인데, 70kg과 다르지 않다고 오판함.")
    print("   (이것이 바로 제2종 오류입니다.)")
else:
    print("결과: 귀무가설 기각")
```

```python
import random

# "이 동전은 앞면이 나올 확률이 50%야"라고 미리 정함 (Design)
expected_prob = 0.5 
trial_count = 10 # 10번 던지기
heads = 0

print(f"--- 실험 시작 (예상 확률: {expected_prob * 100}%) ---")

for i in range(trial_count):
    # 실제 동전 던지기 (0~1 사이 랜덤 숫자 뽑기)
    if random.random() < expected_prob:
        heads += 1
        result = "앞면"
    else:
        result = "뒷면"
    print(f"{i+1}번째 결과: {result}")

# --- 3. 상대도수 (Relative Frequency): 실제 결과 ---
actual_freq = heads / trial_count

print("-" * 30)
print(f"우리의 예상 (확률): {expected_prob * 100}%")
print(f"실제 결과 (상대도수): {actual_freq * 100}%")
```

```python
import math

# 1. 설정값 정의
total_balls = 7
white_balls = 4
black_balls = 3
pick_count = 3

# 2. 전체 경우의 수: C(7, 3)
total_cases = math.comb(total_balls, pick_count)

# 3. 사건의 경우의 수: C(4, 2) x C(3, 1)
# 흰공 2개 뽑기 x 검은공 1개 뽑기
event_cases = math.comb(white_balls, 2) * math.comb(black_balls, 1)

# 4. 확률 계산
probability = event_cases / total_cases

print(f"계산된 확률 (분수): {event_cases}/{total_cases}")
print(f"계산된 확률 (소수): {probability:.4f}")
```

```python
import random

def run_simulation(trials):
    # 1. 주머니 만들기 (리스트로 구현)
    # 흰공(W) 4개, 검은공(B) 3개
    urn = ['W', 'W', 'W', 'W', 'B', 'B', 'B','B']
    
    success_count = 0
    
    # 2. trials 횟수만큼 반복 실험 (for문)
    for _ in range(trials):
        # 주머니에서 공 3개를 비복원 추출 (뽑은 것 다시 안 넣음)
        drawn_balls = random.sample(urn, 6)
        
        # 3. 조건 확인: 흰공(W)이 2개이고, 검은공(B)이 1개인가?
        if drawn_balls.count('W') == 3 and drawn_balls.count('B') == 3:
            success_count += 1
            
    # 4. 상대도수 계산
    return success_count / trials

# --- 실행 ---
#trials = 100000  # 10만 번 실험
trials=100
simulated_prob = run_simulation(trials)

print(f"시뮬레이션 횟수: {trials}번")
print(f"실험 결과 확률: {simulated_prob:.4f}")
print(f"이론적 확률(0.5143)과의 오차: {abs(simulated_prob - 18/35):.5f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

def monte_carlo_pi(n_samples):
    # 1. 0과 1 사이의 무작위 x, y 좌표 생성
    x = np.random.rand(n_samples)
    y = np.random.rand(n_samples)
    
    # 2. 원점(0,0)으로부터의 거리 계산 (x^2 + y^2 <= 1 인지 확인)
    dist = x**2 + y**2
    
    # 3. 원 안에 들어온 점들 분류
    inside_circle = dist <= 1
    pi_estimate = (np.sum(inside_circle) / n_samples) * 4
    
    return x, y, inside_circle, pi_estimate

# 10,000번 시뮬레이션 실행
n = 10000
x, y, inside, pi_val = monte_carlo_pi(n)

print(f"시뮬레이션 결과 추정된 PI 값: {pi_val}")

# 시각화
plt.figure(figsize=(6, 6))
plt.scatter(x[inside], y[inside], color='blue', s=1, label='Inside')
plt.scatter(x[~inside], y[~inside], color='red', s=1, label='Outside')
plt.legend()
plt.title(f"Monte Carlo Simulation (n={n}, Pi={pi_val})")
plt.show()
```

```python
import random

# --- Case 1: 독립사건 (주사위와 동전) ---
# 서로 영향을 주지 않으므로 각각 정의하고 그냥 곱합니다.
prob_dice_even = 0.5  # 주사위 짝수
prob_coin_head = 0.5  # 동전 앞면

# P(A ∩ B) = P(A) * P(B)
prob_independent = prob_dice_even * prob_coin_head
print(f"독립사건 확률: {prob_independent} (0.25)")


# --- Case 2: 종속사건 (제비뽑기: 당첨 2, 꽝 8) ---
# 앞사람이 뽑은 결과가 뒷사람 확률을 바꿉니다.
total = 10
win = 2

# A: 첫 번째 사람이 당첨될 확률 P(A)
prob_A = win / total  # 2/10

# B|A: (A가 당첨된 후) 두 번째 사람도 당첨될 확률 P(B|A)
# 전체도 1 줄고, 당첨제비도 1 줄어듭니다. (상태 변경!)
prob_B_given_A = (win - 1) / (total - 1) # 1/9

# P(A ∩ B) = P(A) * P(B|A)
prob_dependent = prob_A * prob_B_given_A
print(f"종속사건 확률: {prob_dependent:.3f} (약 0.022)")
```
