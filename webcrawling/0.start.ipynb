```python
import requests
from bs4 import BeautifulSoup

# 1. URL 설정
url = "https://finance.naver.com/marketindex/"

# 2. 크롬 브라우저의 User-Agent 설정
# 실제 크롬 브라우저에서 접속하는 것처럼 보이게 합니다.
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/121.0.0.0 Safari/537.36"
}

try:
    # 3. HTML 데이터 가져오기
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.text, "html.parser")

    # 4. 데이터 추출 (크롬 개발자 도구로 확인한 선택자 활용)
    # '미국 USD' 환율 값은 '.value' 클래스 안에 들어있습니다.
    exchange_rate = soup.select_one("#exchangeList > li.on > a.head.usd > div > span.value").text
    
    # 5. 결과 출력
    print(f"현재 원/달러 환율: {exchange_rate}원")

except Exception as e:
    print(f"오류가 발생했습니다: {e}")
```

```python
import requests
from bs4 import BeautifulSoup

# 1. 대상 URL (iframe 주소)
url = "https://finance.naver.com/marketindex/exchangeList.naver"

# 2. 크롬 브라우저 User-Agent 설정
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/122.0.0.0 Safari/537.36"
}

# 3. OECD 선진국 필터 리스트
oecd_countries = ['미국', '유럽연합', '일본']

try:
    # 데이터 가져오기
    response = requests.get(url, headers=headers)
    
    # [수정] 인코딩을 cp949로 강제 설정하여 한글 깨짐 방지
    response.encoding = 'cp949' 
    
    soup = BeautifulSoup(response.text, "html.parser")

    # 테이블의 모든 행(tr) 추출
    rows = soup.select("table.tbl_exchange tbody tr")

    print(f"{'국가(통화)':<15} | {'매매기준율':<10} | {'등락'}")
    print("-" * 45)

    for row in rows:
        # 각 항목 추출 시 존재 여부 확인 (에러 방지)
        tit_tag = row.select_one("td.tit")
        sale_tag = row.select_one("td.sale")
        
        if not tit_tag or not sale_tag:
            continue
            
        name = tit_tag.text.strip()
        rate = sale_tag.text.strip()
        
        # OECD 국가인지 필터링
        if any(country in name for country in oecd_countries):
            # 등락 이미지(상승/하락)와 수치 가져오기
            img_tag = row.select_one("td:nth-child(3) img")
            direction = img_tag['alt'] if img_tag else "보합"
            
            change_tag = row.select_one("td.total")
            change = change_tag.text.strip() if change_tag else "0.00"

            print(f"{name:<12} | {rate:>10} | {direction} {change}")

except Exception as e:
    print(f"❌ 실행 중 오류가 발생했습니다: {e}")
```

```python
import requests
from bs4 import BeautifulSoup

# 1. 전체 환율 데이터가 있는 실제 주소 (iframe 소스)
url = "https://finance.naver.com/marketindex/exchangeList.naver"
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/121.0.0.0 Safari/537.36"
}

# 2. OECD 주요 회원국 필터링 리스트 (네이버 표기 기준)
oecd_countries = [
    '미국', '유럽연합', '일본', '영국', '스위스', '캐나다', '호주', '뉴질랜드', 
    '체코', '덴마크', '노르웨이', '스웨덴', '헝가리', '폴란드', '멕시코', 
    '튀르키예', '이스라엘', '칠레', '콜롬비아'
]

try:
    response = requests.get(url, headers=headers)
    response.encoding = 'euc-kr'  # 네이버 금융은 euc-kr을 사용합니다.
    soup = BeautifulSoup(response.text, "html.parser")

    # 모든 환율 행(tr) 가져오기
    rows = soup.select("table.tbl_exchange tbody tr")

    print(f"{'국가/통화':<15} | {'매매기준율':<10}")
    print("-" * 30)

    for row in rows:
        # 통화명 추출 (예: '미국 USD')
        name = row.select_one("td.tit").text.strip()
        # 환율 값 추출
        rate = row.select_one("td.sale").text.strip()

        # OECD 필터링: 국가명이 리스트에 포함되어 있는지 확인
        if any(country in name for country in oecd_countries):
            print(f"{name:<15} | {rate:>10}")

except Exception as e:
    print(f"오류가 발생했습니다: {e}")
```
